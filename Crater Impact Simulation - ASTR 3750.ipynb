```python
import numpy as np
from random import randint
from math import isclose
import matplotlib.pyplot as plt
%matplotlib inline
grid = np.full((50,50), '-')
shapes = ['x', ' ']
saturated  = False
time_count = 0 
time_array = []
amt_craters_time = []
t_sat_25 = 0
t_sat_50 = 0
t_sat_75 = 0
```

```python
def how_many_craters(grid):
    count_craters = np.count_nonzero(grid == 'o')
    amt_craters_time.append(count_craters)
    return count_craters
def is_saturated(grid):
    count = np.count_nonzero(grid == '-')
    if count < 0.05*2500:
        saturated = True
        return saturated
    else:
        saturated = False
        return saturated
```

```python
print(f'% Saturation ~= 0%               number of craters: 0     time: {time_count} in thousands of years')
for x in grid:
    print(' '.join(x))
print()
print()
while saturated == False:
    saturated = is_saturated(grid)
    time_count +=1
    time_array.append(time_count)
    x = randint(2, 47)
    y = randint(2, 47)
    a = np.full((5,5), 'x')
    a[1:4, 1:4] = ' '
    a[2,2] = 'o'
    for e in shapes:
        if e in grid[x-2:x+3, y-2:y+3]:
            a[1:4, 1:4] = ' '
            a[2,2] = 'o'
            a[0,0] = ' '
            a[0,4] = ' '
            a[4,4] = ' '
            a[4,0] = ' '
        else:
            a[1:4, 1:4] = ' '
            a[2,2] = 'o'
            a[0,0] = '-'
            a[0,4] = '-'
            a[4,4] = '-'
            a[4,0] = '-'
    grid[x-2:x+3, y-2:y+3] = a
    crater_num = how_many_craters(grid)
    if np.count_nonzero(grid == '-') > 1860 and np.count_nonzero(grid == '-') < 1885 and t_sat_25 < 1:
        time25 = time_count
        t_sat_25 += 1
        print(f'% Saturation ~= 25%               number of craters: {crater_num}     time: {time_count} in thousands of years')
        for x in grid:
            print(' '.join(x))
        print()
        print()
    elif np.count_nonzero(grid == '-') > 1240 and np.count_nonzero(grid == '-') < 1255 and t_sat_50 < 1:
        time50 = time_count
        t_sat_50 += 1
        print(f'% Saturation ~= 50%               number of craters: {crater_num}     time: {time_count} in thousands of years')
        for x in grid:
            print(' '.join(x))
        print()
        print()
    elif np.count_nonzero(grid == '-') > 620 and np.count_nonzero(grid == '-') < 635 and t_sat_75 < 1:
        time75 = time_count
        t_sat_75 += 1
        print(f'% Saturation ~= 75%               number of craters: {crater_num}     time: {time_count} in thousands of years')
        for x in grid:
            print(' '.join(x))
        print()
        print() 
print((f'% Saturation ~= 100%               number of craters: {crater_num}     time: {time_count} in thousands of years'))
for x in grid:
    print(' '.join(x))
```

**c**

(10 pts) Plot the number of craters evident on the surface as a function of time, and
indicate the time to saturation.

```python
plt.figure(figsize = (20,8))
plt.title(f'Amount of craters with respect to time. Time to saturation: {time_count*10**3} years')
plt.ylabel('Amount of craters'), plt.xlabel('Time in thousands of years')
plt.axvline(0, linestyle = '--', c = 'green', label = '0%'), plt.axvline(time25, linestyle = '--', c = 'red', label = '25%'), plt.axvline(time50, linestyle = '--', c = 'blue', label = '50%'), plt.axvline(time75, linestyle = '--', c = 'purple', label = '75%'), plt.axvline(time_count, linestyle = '--', c = 'black', label = '100%')
plt.plot(time_array, amt_craters_time)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20);
```
